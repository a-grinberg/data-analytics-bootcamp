### Exercise 1 : Linear Regression Analysis

**Instructions**

Objective:
Apply linear regression to a dataset and interpret the results.

Task:

Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.

Dataset:

house_sizes: [50, 70, 80, 100, 120] (in square meters) \
house_prices: [150,000, 200,000, 210,000, 250,000, 280,000] (in currency units)

Questions:

- What is the slope and intercept of the regression line?
- Predict the price of a house that is 90 square meters.
- Interpret the meaning of the slope in the context of housing prices.

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

slope, intercept, r, p, std_err = stats.linregress(house_sizes, house_prices)

print("Slope:", slope)
print("Intercept:", intercept)
predicted_price = intercept + slope * 90
print("Predicted price of a house with 90 square meters:", predicted_price)

plt.scatter(house_sizes, house_prices)
plt.plot(house_sizes, intercept + slope * house_sizes, 'r')
plt.xlabel('House Size (square meters)')
plt.ylabel('House Price (currency units)')
plt.show()
````

As we can see, the slope of the regression line represents the average increase in house price for each additional square meter of house size

### Exercise 2 : Understanding ANOVA

Instructions

*Objective:* Test understanding of ANOVA and its application.

*Task:* Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.

Dataset:

fertilizer_1: [5, 6, 7, 6, 5] \
fertilizer_2: [7, 8, 7, 9, 8] \
fertilizer_3: [4, 5, 4, 3, 4] \

Questions:

- Perform an ANOVA test on the given data. What are the F-value and P-value?
- Based on the P-value, do the fertilizers have significantly different effects on plant growth?
- Explain what would happen if the P-value were greater than 0.05.

````python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("F-value:", f_value)
print("P-value:", p_value)
````

- If the P-value were greater than 0.05, we would fail to reject the null hypothesis.
This would mean that there is not enough evidence to suggest that the fertilizers have significantly different effects on plant growth.
In other words, any observed differences in growth between the groups could be due to random chance.

### Exercise 4 : Correlation Coefficients

**Instructions**

Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.

Example in code:


```
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]}
```


````python
import pandas as pd
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
pearson_corr = data['age'].corr(data['income'], method='pearson')
spearman_corr = data['age'].corr(data['income'], method='spearman')
print("Pearson correlation coefficient:", pearson_corr)
print("Spearman correlation coefficient:", spearman_corr)
````
