# Exercise 5 : Export A Dataframe To Excel Format And JSON Format.

- Create a simple dataframe.
- Export the dataframe to an excel file.
- Export the dataframe to a JSON file.

```python
!pip install faker
```

```python
from faker import Faker
import pandas as pd

instance = Faker()

df_ex5 = pd.DataFrame([[instance.first_name(), instance.last_name(), instance.email()] for i in range(100)], columns=['first_name', 'last_name', 'email'])
df_ex5.to_excel('ex5.xlsx')
df_ex5.to_json('ex5.json', orient='records')
```
